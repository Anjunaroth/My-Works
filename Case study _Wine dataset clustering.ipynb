```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df= scaler.fit_transform(data)
```

## Making Clusters

### 1)K_means

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init= 'k-means++',random_state=42)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
    print(kmeans.inertia_)
```

```python
fig = plt.figure(figsize=(10,6))
plt.plot(range(1,11),wcss,'-o')
plt.xlabel('No of clusters')
plt.ylabel('WCSS')
plt.title('The elbow method')
plt.show()
```

From Elbow Method it is Clear that the optimum no.of Clusters is 3.

```python
kmeans = KMeans(n_clusters=3, init= 'k-means++',random_state=42)
y_kmeans=kmeans.fit_predict(df)
y_kmeans
```

```python
data['K_means_Cluster'] =y_kmeans
data.head()
```

```python
df1 = data[data['K_means_Cluster']==0]
df2 = data[data['K_means_Cluster']==1]
df3 = data[data['K_means_Cluster']==2]

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Alcohol', y='Ash_Alcanity', data=df1, color='green')
sns.scatterplot(x='Alcohol', y='Ash_Alcanity', data=df2, color='red')
sns.scatterplot(x='Alcohol', y='Ash_Alcanity', data=df3, color='blue')
plt.xlabel('Alcohol')
plt.ylabel('Ash_Alcanity')
plt.show()
```

```python
#Checking the goodness of fit
from sklearn.metrics import silhouette_score
score = silhouette_score(data, y_kmeans)
score
```

### 2)Agglomerative heirarchical clustering

```python
from scipy.cluster.hierarchy import dendrogram,linkage
linked=linkage(df,method='ward')
plt.figure(figsize=(15,8))
dendrogram(linked,orientation='top',distance_sort='ascending',show_leaf_counts=True)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc= AgglomerativeClustering(n_clusters = 3,affinity= 'euclidean',linkage='ward')
labels= hc.fit_predict(df)
```

```python
labels
```

```python
data['Hierarchical_Cluster'] = labels
data.head()
```

```python
df1 = data[data['Hierarchical_Cluster']==0]
df2 = data[data['Hierarchical_Cluster']==1]
df3 = data[data['Hierarchical_Cluster']==2]
plt.figure(figsize=(8, 6))

sns.scatterplot(x='Alcohol', y='Ash_Alcanity', data=df1, color='green')
sns.scatterplot(x='Alcohol', y='Ash_Alcanity', data=df2, color='red')
sns.scatterplot(x='Alcohol', y='Ash_Alcanity', data=df3, color='blue')
plt.xlabel('Width')
plt.ylabel('Length')
plt.show()
```
