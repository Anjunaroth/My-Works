```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel('Telco-Customer-Churn.xlsx')
```

```python
data
```

### Compare churn count with respect to gender

```python
import seaborn as sns
```

```python
sns.countplot(x='gender',data=data,hue='Churn')
```

### Find out how many female senior citizens there in the datase

```python
data1=data[['gender','SeniorCitizen']]
data1
```

```python
data1.loc[data1['SeniorCitizen'] == 1, 'gender'].value_counts()
```

### Compare tenure with Total Charges

```python
sns.catplot(x = "Contract",      
            y = "TotalCharges",       
            hue = "InternetService",  
            data = data, 
            kind = "bar")
```

### Find out which contract is preferred by the senior citizen

```python
df_SCZN = data[data.SeniorCitizen == 1.0]
df_SCZN
```

```python
df_SCZN.groupby('InternetService').agg('Contract').value_counts().plot(kind='bar', x='Contract', y='Count')
```
