```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Read the dataset to the python environment

```python
data = pd.read_csv('cars_data.csv')
```

### Check for the null values present in the dataset.

### Plot a bar graph of male vs female buyers participated in the sales

```python
data.loc[data['Buyer Gender'] == 'Female', 'Buyer Gender'].value_counts()
```

```python
data.loc[data['Buyer Gender'] == 'Male', 'Buyer Gender'].value_counts()
```

```python
sns.countplot(x="Buyer Gender", data=data)
```

```python
data.groupby('Color').agg('Buyer Gender').value_counts().plot(kind='bar', x='Buyer Gender', y='Count')
```
